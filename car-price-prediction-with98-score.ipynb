```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("/kaggle/input/car-price-prediction/CarPrice_Assignment.csv")
df
```

```python
df.columns
```

```python
inputs = df.drop(['car_ID','symboling','aspiration','doornumber','carbody','drivewheel', 'enginelocation', 'wheelbase','carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype','cylindernumber', 'fuelsystem', 'boreratio', 'stroke','compressionratio','peakrpm', 'citympg', 'highwaympg','price'],axis='columns')
inputs
```

```python
inputs.fueltype = inputs.fueltype.map({'gas': 1, 'diesel': 0})
inputs
```

```python
dummies = pd.get_dummies(df.CarName)
dummies
```

```python
merged = pd.concat([inputs,dummies],axis='columns')
merged
```

```python
final = merged.drop(['CarName'], axis='columns')
final
```

```python
target = df.price
target
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(final,target)
```

```python
model.predict(final)
```

```python
model.score(final,target)
```
